## Seoul Sector

```python
#import custom module
import nre.classes as nc
import nre.util as nu
import pickle
from tqdm import tqdm_notebook

nu.IS_LOGGING = False
```

```python
regions = nu.get_region_list('0000000000')
for i, region in enumerate(regions): 
    print(i, region)
idx = int(input("원하는 도시 인덱스 : "))
cur_region = regions[idx]

print(cur_region)

regions = nu.get_region_list(cur_region.no)

region_sector = []

for child in regions:
    region_sector.extend(nu.get_region_list(child.no))
    
print('Found', len(region_sector))

for se in region_sector:
    print(se.name)
```

## Region To Sector

```python
# region to sector
seq_regions = nu.split_list(region_sector)

print(len(seq_regions))

sectors = []
cancels = []

for regs in seq_regions:
    nu.get_sleep(50)
    se, cn = nu.get_sector_list(regs, loop=tqdm_notebook)
    sectors.extend(se)
    cancels.extend(cn)

with open('%s sectors.data' % cur_region.name, 'wb') as f:
    pickle.dump(sectors, f)
```

```python
sectors = None

with open('seoul sectors.data', 'rb') as f:
    sectors = pickle.load(f)

print('Sector', len(sectors))
```

```python
import pandas as pd

def old_classifier(x): #old 구분
    value = abs(int(x[0:4]) - 2022)
    if value <= 5:
        return 'New'
    elif value <= 10:
        return 'Common'
    return 'Old'
    
def get_df(sector, things, neighbors, standard):
    nu.update_things_intersection(things, neighbors, nu.get_distance_standard(standard))
    total_df = pd.DataFrame(data=[t.get_list() for t in things], columns=nc.NThing.HEADER)
    total_df['Old'] = total_df['Build'].apply(old_classifier)
    total_df['sector'] = sector.name
    total_df['divisition'] = sector.divisition
    total_df['city'] = sector.city
    total_df.to_csv('./csv/%s - %s house_price.csv' % (sector.name, sector.city), sep=',', na_rep='NaN', index=False)
    
```

```python
error_sector = []
infos = []

for sec in tqdm_notebook(sectors):
    try:
        bundle = nu.get_all_on_sector(sec)
        infos.append(bundle)
    except:
        error_sector.append(sec)
        print('Error', sec)
```

```python
with open('sectors bundles.data', 'wb') as f:
    pickle.dump(infos, f)
    
with open('fail sectors.data', 'wb') as f:
    pickle.dump(error_sector, f)
```
