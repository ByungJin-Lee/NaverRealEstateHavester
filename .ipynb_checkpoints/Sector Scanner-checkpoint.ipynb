```python
#서울 지역 Sector 검색
seoul_region = nc.NRegion('서울시', None, '1100000000')

seoul_child = nu.get_region_list(seoul_region.no)

region_sector = []

for child in seoul_child:
    region_sector.extend(nu.get_region_list(child.no))
    
print('Found', len(region_sector))

for se in region_sector:
    print(se.name)
```

```python
import pandas as pd

def old_classifier(x): #old 구분
    value = abs(int(x[0:4]) - 2022)
    if value <= 5:
        return 'New'
    elif value <= 10:
        return 'Common'
    return 'Old'

def get_df_save(sector):
    things = nu.get_things_each_direction(sector)
    neighbors = nu.get_all_neighbors(sector)
    nu.update_things_intersection(things, neighbors, nu.get_distance_standard())
    total_df = pd.DataFrame(data=[t.get_list() for t in things], columns=nc.NThing.HEADER)
    total_df['Old'] = total_df['Build'].apply(old_classifier)
    total_df['sector'] = sector.name
    total_df['city'] = sector.city
    total_df.to_csv('./csv/%s - %s house_price.csv' % (sector.name, sector.city), sep=',', na_rep='NaN', index=False)
    
```

```python
error_sector = []

for sec in tqdm_notebook(sectors):
    try:
        get_df_save(sec)
    except:
        error_sector.append(sec)
        print('Error', sec)

# get_df_save(sectors[0])
# import pickle

# with open('seoul sectors.data', 'wb') as f:
#     pickle.dump(sectors, f)
```
