## Seoul Sector

```python
#import custom module
import nre.con as nc
import nre.util as nu
from tqdm import tqdm_notebook

nu.IS_LOGGING = False
```

```python
#서울 지역 Sector 검색
seoul_region = nc.NRegion('서울시', None, '1100000000')

seoul_child = nu.get_region_list(seoul_region.no)

region_sector = []

for child in seoul_child:
    region_sector.extend(nu.get_region_list(child.no))
    
print('Found', len(region_sector))

for se in region_sector:
    print(se.name)
```

## Region To Sector

```python
#region to sector
seq_regions = nu.split_list(region_sector)

print(len(seq_regions))

sectors = []
cancels = []

for regs in seq_regions:
    nu.get_sleep(60)
    se, cn = nu.get_sector_list(regs, loop=tqdm_notebook)
    sectors.extend(se)
    cancels.extend(cn)

print('Sector', len(sectors))
```

```python
import pandas as pd

def old_classifier(x): #old 구분
    value = abs(int(x[0:4]) - 2022)
    if value <= 5:
        return 'New'
    elif value <= 10:
        return 'Common'
    return 'Old'

def get_things(sector):
    smap = sector.map
    addon = nc.NAddon(
        #direction=nc.NAddon.DIR_EACH, #전 방향 탐색의 경우 사용 - 아래 로직을 바꾸세요
        tradeType=[nc.NAddon.TRADE_DEAL, nc.NAddon.TRADE_LEASE], #목표 거래 - 매매, 전세
        estateType=[nc.NAddon.ESTATE_APT, nc.NAddon.ESTATE_OPST] #목표 매물 - 아파트, 오피스텔
    )
    things = [] # 매물 기록
    for dirr in nc.NAddon.DIR_EACH: # 모든 방향 (남향 등등)
        addon.dir = dirr # 방향 조건 선택
        values = []
        for v in nu.get_things(sector, addon):
            if smap.contain(v.loc) is False: # 구역 포함되어있는지
                continue 
            v.dir = dirr # 매물 방향 정의
            values.append(v)
        things.extend(values)
    return things

def get_nei(sector):
    smap = sector.map
    neighbors = [] # 편의시설 기록
    for nType in nc.NNeighbor.EACH: # 모든 편의시설
        values = []
        res = nu.get_neighborhood(sector.loc, nType)
        for r in res:
            if smap.contain(r.loc):
                values.append(r)
        neighbors.extend(values)
    return neighbors
    
def things_and_nei(things, neighbors, standard):
    for thing in things: # 매물
        around = nc.NNeighborAround()
        for nei in neighbors: # 편의시설
            d = nu.distance_between(thing.loc, nei.loc)
            if d <= standard[nei.type]: # meter
                around.increase(nei.type)
        thing.neiAround = around

def get_df(sector):
    things = get_things(sector)
    nei = get_nei(sector)
    things_and_nei(things, nei, nu.get_distance_standard())
    thing_df = pd.DataFrame(data=[t.get_list() for t in things], columns=nc.NThing.HEADER)
    nei_df = pd.DataFrame(data=[t.neiAround.get_list() for t in things], columns=nc.NNeighborAround.HEADER)
    total_df = pd.concat([thing_df, nei_df], axis=1)
    total_df['Old'] = total_df['Build'].apply(old_classifier)
    total_df['sector'] = sector.name
    return total_df
```

```python
cancelss = []

seq_sectors = nu.split_list(sectors)

df = get_df(sectors[0])
df.to_csv('./csv/%s - %s house_price.csv' % (sectors[0].name, sectors[0].city), sep=',', na_rep='NaN', index=False)
```
